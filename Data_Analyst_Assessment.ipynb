```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
merged_all = pd.merge(pd.merge(df2, df1, on=['Student_ID', 'Course_ID'], how='inner'),
                      df, on='Student_ID', how='inner')
```

```python
merged_all['Date'] = pd.to_datetime(merged_all['Date'])
```

```python
sns.set(style="whitegrid")
```

```python
# 1. Engagement pattern by Course
plt.figure(figsize=(10, 6))
course_engagement = merged_all.groupby('Course_ID')['Time_Spent_Minutes'].mean().sort_values(ascending=False)
sns.barplot(x=course_engagement.values, y=course_engagement.index, palette="viridis")
plt.title('Average Engagement (Time Spent) per Course')
plt.xlabel('Average Time Spent (Minutes)')
plt.ylabel('Course ID')
plt.tight_layout()
plt.show()
```

```python
# 2. Engagement by Age Group
plt.figure(figsize=(10, 6))
merged_all['Age_Group'] = pd.cut(merged_all['Age'], bins=[15, 20, 25, 30, 35, 40], labels=['15-20', '21-25', '26-30', '31-35', '36-40'])
age_group_engagement = merged_all.groupby('Age_Group')['Time_Spent_Minutes'].mean()
sns.barplot(x=age_group_engagement.index, y=age_group_engagement.values, palette="coolwarm")
plt.title('Engagement by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Time Spent (Minutes)')
plt.tight_layout()
plt.show()
```

```python
# 3. Feedback Ratings by Course
plt.figure(figsize=(10, 6))
avg_rating_course = merged_all.groupby('Course_ID')['Rating'].mean().sort_values(ascending=False)
sns.barplot(x=avg_rating_course.values, y=avg_rating_course.index, palette="crest")
plt.title('Average Feedback Rating per Course')
plt.xlabel('Average Rating')
plt.ylabel('Course ID')
plt.tight_layout()
plt.show()
```

```python
# 4. Heatmap: Engagement by Location and Course
heatmap_data = merged_all.groupby(['Location', 'Course_ID'])['Time_Spent_Minutes'].mean().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, annot=True, fmt=".1f")
plt.title('Engagement Heatmap (Location vs Course)')
plt.xlabel('Course ID')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

```python
# 5. Trend: Average Engagement Over Time
time_trend = merged_all.groupby('Date')['Time_Spent_Minutes'].mean()
plt.figure(figsize=(12, 6))
time_trend.plot(marker='o', linestyle='-', color='teal')
plt.title('Trend of Average Engagement Over Time')
plt.xlabel('Date')
plt.ylabel('Average Time Spent (Minutes)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
